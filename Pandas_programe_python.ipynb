## Excluindo linhas e colunas com NaN

df.dropna() Exclui todas as linhas que tenham valores missing
Por padrão ela dropa axis = 0 que analisa linhas

df.dropna(how = 'any') -> qualquer linha que tenha missing

df.dropna(how = 'all') -> exclui se a linha ou coluna tiver todos os valores missing

df.dropna(thresh = 1) -> exclui linhas que não tem pelo menos um valor normal ou conforme for passado por parâmetro

pd.set_option('float_format','{:.3f}.format') -> formata o número ou sequência de número, 3 casas depois do ponto


## Excluindo linhas e colunas

df.drop('col1', axis=1) -> apaga coluna

df.drop(0, axis=0) -> apaga linha

df.drop(['col1', 'col2'], axis=1) -> apagando duas colunas

df.drop(['col1', 'col2'], axis=0, errors='ignore') -> ignorando erros

df.drop(['col1', 'col2'], axis=1, inplace=True) -> modificação feita diretamente do dataframe

## Alterando coluna baseada em outra



```python
dados7 = [['C', 6], ['C', 15], ['C', 1], ['V', 19], ['V', 20],
          ['C', 18], ['V', 8], ['V', 1], ['C', 9]]
df = pd.DataFrame(dados7, columns=['col1', 'col2'])
df
```

```python
#UMA CONDIÇÃO
df.loc[df['col1'] == 'V', 'col2'] = -df['col2']
df
```

```python
#DUAS CONDIÇÕES
df.loc[(df['col1'] == 'V') & (df['col2'] > 10), 'col2'] = -df['col2']
df
```

```python
df.loc[df['col2'] > 10, 'maior_que_10'] = 'sim'
df.loc[df['col2'] <= 10, 'maior_que_10'] = 'nao' #preencher os valores nulos com NÃO
# ou df.fillna('não')
df
```
